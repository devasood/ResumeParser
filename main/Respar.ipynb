```python
import pandas as pd
import numpy as np

def convert_nan(style,the_value="Normal"):
    if type(style)==float:
        return the_value
    return style

def remove_start_quotes(face):
    if face[0]=="\'":
        return face[1:]
    return face

def remove_end_quotes(face):
    if face[-1]=="\'":
        return face[:-1]
    return face
    
def read_words():
    df = pd.read_csv("input.txt",delimiter=" ",usecols=[1,2,3,4,5,6,7,8])

    df["FontFace"] = df["FontFace"].apply(remove_start_quotes)
    df["FontFace"] = df["FontFace"].apply(remove_end_quotes)
    df["FontStyle"] = df["FontStyle"].apply(convert_nan)
    df["FontStyle"] = df["FontStyle"].apply(remove_end_quotes)
    df["Letter"] = df["Letter"].apply(lambda letter:convert_nan(letter," "))
    df = df.copy()[["FontFace","FontStyle","FontSize","Letter","ymaxpos"]]
#     return df.set_index(["FontFace","FontSize","FontStyle"])
    return df

read_words()
```

```python
def get_words():
    df=read_words()
    data = df.as_matrix().tolist()
    
    chunks=[]
    block=[]
#     count=0
    for index,row in enumerate(data):
        if(row[3]==" "):
            
            if(data[index+1][0] != data[index-1][0]):
                block = "".join(block).strip()
                if len(block)!=0:
                    chunks.append([block])
                block=[]
            else:
                block.append(" ")

#             count=count+1
#             if(count==10):
#                 break
        else:
            block.append(row[3])
    
    return chunks

get_words()
```
